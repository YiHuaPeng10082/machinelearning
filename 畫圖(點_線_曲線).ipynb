```python
#隨機畫出50個點
import numpy as np
import matplotlib.pyplot as pt #matplotlib 畫圖
N=50
x=np.random.rand(N) #亂數產生
y=np.random.rand(N)
pt.scatter(x,y) #不重複
pt.show()
```

```python
#sin 
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/180.0)
pt.xlim(0,360)   #左右X軸的距離
pt.ylim(-1.1,1.1) #上下Y軸的距離
pt.title("Sin Title") #顯示標題
pt.plot(x,y) #連續的
pt.show()
```

```python
#sin and cos
import numpy as np
import matplotlib.pyplot as pt
x=np.arange(0,360)
y=np.sin(x*np.pi/60.0)#頻率
z=np.cos(x*np.pi/60.0)#頻率
pt.xlim(0,360)   #左右X軸的距離
pt.ylim(-1.02,1.02) #上下Y軸的距離
pt.title("Sin and Cos") #顯示標題
pt.plot(x,y,color="Green") #指定顏色
pt.plot(x,z,color="Red")
pt.show()
```

```python
#直線
import numpy as np
import matplotlib.pyplot as pt
x=[1,2,3,4]
pt.title("straight line") #顯示標題
pt.plot(x) #連續的
pt.show()
```

```python
#直線
import numpy as np
import matplotlib.pyplot as pt
x=[1,2,3,4]
y=[5,6,8,2]
pt.title("straight line") #顯示標題
pt.plot(x,y) #連續的
pt.xlabel("X")#X軸標籤
pt.ylabel("Y")#Y軸標籤
pt.show()
```

```python
#虛線
import numpy as np
import matplotlib.pyplot as pt
x=[1,2,3,4]
pt.title("straight line") #顯示標題
pt.plot(x,"r--") # "r--"虛線
pt.show()
```

```python
#bar
import numpy as np
import matplotlib.pyplot as pt
x=[1,2,3,4]
y=[2,10,3,8]
pt.title("straight line") #顯示標題
pt.bar(x,y,label="bar chart") #長條圖
pt.legend()
pt.show()
```
